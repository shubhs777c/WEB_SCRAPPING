```python
# Import required libraries
import numpy as np
import pandas as pd

# Import additional libraries
from urllib.request import urlopen
from bs4 import BeautifulSoup

# Getting URL of the scraping dataset
url = 'https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_population'

# Read the HTML table from the webpage
data = pd.read_html(url, header=0)

# Choose the specific table you want to work with from the list of DataFrames
# In this case, we want the first table, which contains the population data
df_0= data[0]
# for idx, df in enumerate(data):
#     print(f"Table {idx+1}:")
#     print(df)
#     print("\n")

# Display the DataFrame
print(df_0)
```

```python
# Horizontal Concatenation
concatenated_df = pd.concat([df_0, df_1, df_2, df_4], axis=1)

# Display the concatenated DataFrame
print(concatenated_df)
```
